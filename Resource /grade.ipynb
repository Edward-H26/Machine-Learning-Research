```python
# import packages
import numpy as np

# input homework grades here
hws = [
    100, 100, 100, 100, 100, 
    100, 100, 100, 100, 100
]

# input lab model grades here
lms = [
   100, 100, 100, 100, 100, 
    100, 100, 100, 100, 100
]

# input lab report grades here
lrs = [
    100
]

# input quiz grades here
q01, q02, q03 = 99, 100, 99

# initialize quiz weights
w01, w02, w03 = 15, 15, 15

# apply adaptive quiz weight updates
if q02 > q01:
    w01 = w01 - 5
    w02 = w02 + 5

if q03 > q01:
    w01 = w01 - 5
    w03 = w03 + 5

if q03 > q02:
    w02 = w02 - 5
    w03 = w03 + 5

# calculate homework subscore
hw = np.mean(hws)
hw = np.min([100, hw])

# calculate lab model subscore
lm = np.mean(lms)
lm = np.min([100, lm])

# calculate lab report subscore
lr = np.mean(lrs)

# report intermediate results
print(f"Homework Subscore: {hw}")
print(f"Lab Model Subscore: {lm}")
print(f"Lab Report Subscore: {lr}")
print(f"Quiz 01 Score: {q01}")
print(f"Quiz 02 Score: {q02}")
print(f"Quiz 03 Score: {q03}")
print(f"Quiz 01 Weight: {w01}%")
print(f"Quiz 02 Weight: {w02}%")
print(f"Quiz 03 Weight: {w03}%")

# calculate overall course percentage
w01 = w01 / 100
w02 = w02 / 100
w03 = w03 / 100
percentage = (
    0.25 * hw +
    0.20 * lm +
    0.10 * lr +
    w01 * q01 +
    w02 * q02 +
    w03 * q03
)

# report overall course percentage
print(f"Course Percentage: {percentage}")
```
